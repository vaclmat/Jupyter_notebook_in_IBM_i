# Work with IBM i database

#IBM i connection

```python
import ibm_db
import ibm_db_dbi
conn = ibm_db.connect("*LOCAL","","")
con = ibm_db_dbi.Connection(conn)
```

```python
# Get the server type
server = ibm_db.server_info( conn )
print("Db2 database server name                 : {}" .format(server.DBMS_NAME))
print("Database name                            : {}" .format(server.DB_NAME))
```

#Create schema

```python
if conn:
    create_schema = 'CREATE SCHEMA PYTHONTST'
    results = ibm_db.exec_immediate(conn, create_schema)
    if results:
                rows = ibm_db.num_rows(results)
                print("affected row:", rows)
    else:
                print(ibm_db.stmt_errormsg())
else:
            print ("Connection failed.")
```

#Create table

```python
if conn:
    create_table = 'CREATE TABLE PYTHONTST.ANIMALS (id INTEGER, breed VARCHAR(32), name CHAR(16), weight DECIMAL(7,2))'
    resultt = ibm_db.exec_immediate(conn, create_table)
    if resultt:
                rows = ibm_db.num_rows(resultt)
                print("affected row:", rows)
    else:
                print(ibm_db.stmt_errormsg())
else:
            print ("Connection failed.")
```

```python
if conn:
    sql = "SELECT * FROM STUDENTRSC.STUDENTDB"
    stmt = ibm_db.exec_immediate(conn, sql)
    result = ibm_db.fetch_both(stmt)
    while( result ):
        print (result)
        result = ibm_db.fetch_both(stmt)
```

```python
from prettytable import PrettyTable
x = PrettyTable()
x.field_names = ["Student ID", "First Name", "Last Name", "Gender"]
x.align["First Name"] = "l"
x.align["Last Name"] = "l"

if conn:
    sql = "SELECT *  FROM STUDENTRSC.STUDENTDB"
    stmt = ibm_db.exec_immediate(conn, sql)
    result = ibm_db.fetch_both(stmt)
    while( result ):
        x.add_row([result[0], result[1].rstrip(), result[2].rstrip(), result[3]])
        result = ibm_db.fetch_both(stmt)
        
print(x)
```

#Insert

```python
new_record = {'Student ID':"I78912345", 'First Name':"aaa", 'last Name':"bbb", 'Gender Type':"male"}
insert_sql = "INSERT INTO STUDENTRSC.STUDENTDB VALUES (?,?,?,?)"
stmt_insert = ibm_db.prepare(conn, insert_sql)
try:
        ibm_db.execute(stmt_insert, tuple(new_record.values()))
except:
        print("Transaction couldn't be completed:" , ibm_db.stmt_errormsg())
else:
        print("sucessfully inserted ")
```
