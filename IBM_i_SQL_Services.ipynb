```python
from prettytable import PrettyTable
x = PrettyTable()
x.field_names = ["JOB_NAME", "AUTHORIZATION_NAME", "ELAPSED_TOTAL_DISK_IO_COUNT", "ELAPSED_CPU_PERCENTAGE"]
x.align["JOB_NAME"] = "l"
x.align["AUTHORIZATION_NAME"] = "l"
if conn:
    sql = "SELECT JOB_NAME, AUTHORIZATION_NAME, ELAPSED_TOTAL_DISK_IO_COUNT, ELAPSED_CPU_PERCENTAGE FROM TABLE(QSYS2.ACTIVE_JOB_INFO(JOB_NAME_FILTER => 'QZDASOINIT', SUBSYSTEM_LIST_FILTER => 'QUSRWRK')) X ORDER BY ELAPSED_TOTAL_DISK_IO_COUNT DESC FETCH FIRST 10 ROWS ONLY"
    stmt = ibm_db.exec_immediate(conn, sql)
    result = ibm_db.fetch_both(stmt)
    while( result ):
        x.add_row([result[0], result[1], result[2], result[3]])
        result = ibm_db.fetch_both(stmt)
        
print(x)
```
